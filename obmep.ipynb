```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python


url_prata = "https://premiacao.obmep.org.br/18obmep/verRelatorioPremiadosPrata-AL.do.htm"

# Realiza a requisição HTTP
response = requests.get(url_prata, verify=False)

# Verifica se a requisição foi bem-sucedida (código 200)
if response.status_code == 200:
    # Parseia o HTML usando o BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrar a tabela no HTML (ajuste de acordo com a estrutura da página)
    # table = soup.find('table', {'class': 'list'})
    table = soup.find_all('table')[2]

    # Usar a função read_html do pan
    # Usar a função read_html do pandas para extrair a tabela
    df = pd.read_html(str(table))[0]

    # # Exibir a tabela
    print(df)

    # # Salvar a tabela em um arquivo CSV
    df.to_csv('tabela_obmep_prata.csv', index=False)
    print("Tabela salva em 'tabela_obmep.csv'")
else:
    print(f"Falha na requisição. Código de status: {response.status_code}")
```
